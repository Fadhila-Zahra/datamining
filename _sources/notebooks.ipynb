## Data Nominal
	Data nominal merupakan salah satu data yang bersifat kategorik. Data ini diperoleh dari pengukuran nominal, yaitu proses pengelompokan beberapa objek yang berbeda berdasarkan karakteristik tertentu. Data ini hanya memungkinkan setiap objek memiliki satu kategori

	Melakukan pengukuran terhadap data nominal dengan menggukan rumus : 
$$ d(i,j)=\frac{p - m}p $$
$ p $ = jumlah data yang akan dihitung jaraknya<br>
$ m $ = jumlah data yang sama

```python
# Menampilkan data Nominal
phone[["brand_name","model_name","os", "release_date"]].head()
```

```python
# Menghitung jarak data Nominal
def d1(i,j):
    P = 4
    M = 0
    for col in ["brand_name","model_name","os", "release_date"]:
        if phone[col][i] == phone[col][j]:
            M += 1

    return (P-M)/P
```

```python
# Menampilkan hasil perhitungan
print(d1(1,2))
print(d1(1,3))
print(d1(1,4))
```

Melakukan normalisasi data karena tidak konsisten
dengan rumus :

$$ z=\frac{x - u}o $$
$ x $ : Data <br>
$ u $ : Rata-rata <br>
$ o $ : Standar deviasi


```python
# Normalisasi popularity
std1 = 353.482673
mean1 = 612.500000
for col1 in ["popularity"]:
    data1 = phone[col1]
    hasil1 = (data1-mean1)/std1
    print(hasil1)

# Normalisasi best_price
std2 = 8891.836260
mean2 = 7941.206699
for col2 in ["best_price"]:
    data2 = phone[col2]
    hasil2 = (data2-mean2)/std2
    print(hasil2)

# Normalisasi lowest_price
std3 = 8560.959059
mean3 = 7716.018672
for col3 in ["lowest_price"]:
    data3 = phone[col3]
    hasil3 = (data3-mean3)/std3
    print(hasil3)

# Normalisasi highest_price
std4 = 11514.936818
mean4 = 9883.410788
for col4 in ["highest_price"]:
    data4 = phone[col4]
    hasil4 = (data4-mean4)/std4
    print(hasil4)

# Normalisasi sellers_amount
std5 = 20.597006
mean5 = 16.743464
for col5 in ["sellers_amount"]:
    data5 = phone[col5]
    hasil5 = (data5-mean5)/std5
    print(hasil5)

# Normalisasi screen_size
std6 = 1.476991
mean6 = 5.394378
for col6 in ["screen_size"]:
    data6 = phone[col6]
    hasil6 = (data6-mean6)/std6
    print(hasil6)

# Normalisasi memory_size
std7 = 111.922576
mean7 = 95.700059
for col7 in ["memory_size"]:
    data7 = phone[col7]
    hasil7 = (data7-mean7)/std7
    print(hasil7)

# Normalisasi battery_size
std8 = 1668.268774
mean8 = 3608.201812
for col8 in ["battery_size"]:
    data8 = phone[col8]
    hasil8 = (data8-mean8)/std8
    print(hasil8)
```

Melakukan perhitungan menggunakan rumus euclidean distance
Rumus :

$$ d(i,j) = \sqrt(|{x_{i1} - x_{j1}}|^2 + |{x_{i2} - x_{j2}}|^2 + ... + |{x_{ip} - x_{jp}}|^2) $$

```python
# Menghitung jarak Numeric dengan metode Euclidean
def d2(i,j,h):
    result = 0
    for col in [hasil1, hasil2, hasil3, hasil4, hasil5, hasil6, hasil7, hasil8]:
        result = math.pow(abs(col[i] - col[j]), h)
    return result**(1/h)
```

```python
# Hasil perhitungan
print(d2(1,2,2))
print(d2(1,3,2))
print(d2(1,4,2))
```

# Tugas 5 (K-Means Clustering)
Clustering adalah proses untuk mengelompokkan data dalam beberapa cluster atau kelompok sehingga data dalam satu cluster memiliki tingkat kemiripan yang maksimum dan data antar cluster yang berbeda memiliki kemiripan minimum. 
Clustering untuk mengelompokkan data berdasarkan klasifikasi yang sama.

K-means mempunyai kemampuan mengelompokkan data dalam jumlah yang cukup besar dengan waktu komputasi yang cepat dan efisien. K-Means merupakan salah satu algoritma klastering dengan metode partisi (partitioning method) yang berbasis titik pusat (centroid) selain algoritma k-Medoids yang berbasis obyek.

Algoritma k-means yaitu :
1. Langkah 1, menentukan secara acak K titik data sebagai pusat cluster yang disebut centroid.
2. Langkah 2, menandai masing-masing xi masuk ke cluster tertentu, dengan cara menghitung jarak xi ke masing-masing pusat cluster(centroid) dan memasukkan xi anggota pusat cluster tertentu tersebut jika memiliki jarak terdekat.
3. Langkah 3, menentukan pusat cluster baru dengan menghitung rata-rata dari anggota cluster.
4. Langkah 4, ulangan
